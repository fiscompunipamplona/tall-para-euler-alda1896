```python
from pylab import plot, show, legend
from math import exp

vnum=[]
time=[]
vana=[]

dt=1
g=9.8
m=68.1
c=12.5
vi=0

for t in range (21):
    
    vi=vi+((c/m)*vi-g)*dt
    v=(m*g/c)*(-exp(c*t/m)+1)
    
    vnum.append (vi)
    time.append (t)
    vana.append (v)
    
#print(vnum) 
#print(vana)

plot (time,vnum)
plot (time,vana)
show
```
